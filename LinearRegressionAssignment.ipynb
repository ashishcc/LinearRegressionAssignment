### Checking VIF

Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is an extremely important parameter to test our linear model. The formula for calculating `VIF` is:

### $ VIF_i = \frac{1}{1 - {R_i}^2} $

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Dropping the variable and updating the model

As you can see from the summary and the VIF dataframe, some variables are still insignificant. One of these variables is, `semi-furnished` as it has a very high p-value of `0.938`. Let's go ahead and drop this variables

```python
# Dropping highly correlated variables and insignificant variables
X_train = X_train.drop(['atemp'], axis=1)

# Build a third fitted model
X_train_lm = sm.add_constant(X_train)
lr_2 = sm.OLS(y_train, X_train_lm).fit()

# Print the summary of the model
print(lr_2.summary())
```

```python
# p-value, VIF
# Keep low-low
# Drop high-high
# Remove first with High p-value and low VIF
# After that remove Low p-value and high VIF
```

## Step 12: Building RFE model
Recursive feature elimination

```python
np.random.seed(0)
df_train, df_test = train_test_split(day, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
scaler = MinMaxScaler()
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['mnth', 'weekday', 'temp', 'atemp', 'hum', 'windspeed', 'cnt']

# fit: learns xmin, xmax
# transform: x - x-min / xmax-xmin
# fit_transform: 
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train
```

### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=10, step=1)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
rfe.support_
rfe.ranking_
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
```

```python
X_train.columns[~rfe.support_]
```

## Building model using statsmodel, for the detailed statistics

```python
X_train_rfe = X_train[col]
```

```python
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
